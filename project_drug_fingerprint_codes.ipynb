```
# import pandas
import pandas as pd
#reas the tsv file that has information about the drug and gene
#index_col to use the first column as row names instead of data
t = pd.read_csv('/content/intersection_DGIdb_mat.tsv', sep ='\t', index_col=0)
t
#print(type(t))
# to ensure all columns are numeric integers
t_numeric = t.astype(int)
# to convert to numpy array
m = t_numeric.values
```

```
# importing pairwise_distances from the sklearn library
from sklearn.metrics import pairwise_distances
# to calculate jacards similarity
jaccard_sim = 1 - pairwise_distances(m, metric = "jaccard")
print(jaccard_sim.shape)
m =t_numeric.values
js_df = pd.DataFrame(
    jaccard_sim,
    index=t.index,
    columns=t.index
)
js_df
#print(np.__version__)
```

```
js_df.to_csv("drugs_similarity_score.tsv",sep= "\t")
from google.colab import files
files.download("drugs_similarity_score.tsv")
```

```
#for fingerprint data
import pandas as pd
#reas the tsv file that has information about the drug and gene
#index_col to use the first column as row names instead of data
h = pd.read_csv('/content/intersection_Fingerprint_mat.tsv', sep ='\t', index_col=0)
h
#print(type(t))
# to ensure all columns are numeric integers
h_numeric = h.astype(int)
# to convert to numpy array
d = h_numeric.values
```

```
jcd_sim = 1 - pairwise_distances(d, metric = "jaccard")
print(jaccard_sim.shape)
d =h_numeric.values
fp_js_df = pd.DataFrame(
    jcd_sim,
    index=h.index,
    columns=h.index
)
fp_js_df
```

```
fp_js_df.to_csv("fingerprint_similarity_score.tsv",sep= "\t")
from google.colab import files
files.download("fingerprint_similarity_score.tsv")
```

```
#import libraries
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```
# to run pca 2 component
pca = PCA(n_components=2)
# to get results
pca_op = pca.fit_transform(js_df.values)
#convert to data frame
pca_df = pd.DataFrame(
    pca_op,
    index = js_df.index,
    columns = ["PC1", "PC2"]
)

# view data frame
# pca_df

# download pca_df
pca_df.to_csv('pca_df_for_drugs.tsv', sep = '\t')
from google.colab import files
files.download("pca_df_for_drugs.tsv")
```

```
# plot pca
plt.figure(figsize=(10,8))
plt.scatter(pca_df["PC1"], pca_df["PC2"], s=10)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA:Drug Similarity Score")
plt.show()
```

```
# for fingerprint data pca
prca = PCA(n_components=2)
# to get results
prca_rslt = prca.fit_transform(fp_js_df.values)
#convert to data frame
prca_df = pd.DataFrame(
    prca_rslt,
    index = fp_js_df.index,
    columns = ["PC1", "PC2"]
)

# view data frame
# prca_df

# download prca_df
prca_df.to_csv('prca_df_for_fingerprint.tsv', sep = '\t')
from google.colab import files
files.download("prca_df_for_fingerprint.tsv")
```

```
# plot pca
plt.figure(figsize=(10,8))
plt.scatter(prca_df["PC1"], prca_df["PC2"], s=10)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA:Fingerprint Similarity Score")
plt.show()
```
