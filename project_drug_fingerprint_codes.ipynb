```
# import pandas
import pandas as pd
#reas the tsv file that has information about the drug and gene
#index_col to use the first column as row names instead of data
t = pd.read_csv('/content/intersection_DGIdb_mat.tsv', sep ='\t', index_col=0)
t
#print(type(t))
# to ensure all columns are numeric integers
t_numeric = t.astype(int)
# to convert to numpy array
m = t_numeric.values
```

```
# importing pairwise_distances from the sklearn library
from sklearn.metrics import pairwise_distances
# to calculate jacards similarity
jaccard_sim = 1 - pairwise_distances(m, metric = "jaccard")
print(jaccard_sim.shape)
m =t_numeric.values
js_df = pd.DataFrame(
    jaccard_sim,
    index=t.index,
    columns=t.index
)
js_df
#print(np.__version__)
```

```
js_df.to_csv("drugs_similarity_score.tsv",sep= "\t")
from google.colab import files
files.download("drugs_similarity_score.tsv")
```
